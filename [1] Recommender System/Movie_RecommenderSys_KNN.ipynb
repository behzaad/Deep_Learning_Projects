A simple movie recommender system using KNN.

```python
import pandas as pd

from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
path_movies = '/content/movies.csv'

movies = pd.read_csv(path_movies, usecols=['movieId', 'title'])

movies.head(5)
```

```python
path_ratings = '/content/ratings.csv'

ratings = pd.read_csv(path_ratings, usecols=['userId', 'movieId', 'rating'])
ratings.head(5)
```

```python
movies_users = ratings.pivot(index= 'movieId', columns= 'userId', values= 'rating').fillna(0)

movies_mat = csr_matrix(movies_users.values)

```

```python
model = NearestNeighbors(metric = 'cosine', algorithm='auto', n_neighbors= 15)

model.fit(movies_mat)
```

```python
def recommender(movie_name, data, n):

  idx = process.extractOne(movie_name, movies['title'], )[2]

  print('movie selected is: ', movies['title'][idx], ' and the index is ', idx)

  distance, indices = model.kneighbors(data[idx], n_neighbors= n)
  for i in indices.squeeze():
    if i != idx:
      print(movies['title'][i])
```

```python
recommender('Anastasia', movies_mat, 10)
```
